**Reading the CSV File and defining Features and Labels**

```python
df = pd.read_csv("/kaggle/input/students-drop-out-prediction/train.csv")
X, y  = df.drop(['id','label'], axis=1), df['label']
```

**Splitting the dataset into Train and Test Data**

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

**Preprocessing**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.fit_transform(X_val)
```

**Model Fitting**

```python
from xgboost import XGBClassifier
model = XGBClassifier(random_state = 42)
model.fit(X_train, y_train)
```

**Predictions on Validation Data**

```python
y_pred_1 = model.predict(X_val)
```

**Evaluation**

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred_1, y_val))
```

**Predictions on Test data**

```python
data1 = pd.read_csv("/kaggle/input/students-drop-out-prediction/test.csv")
X_test = data1.drop(['id'], axis=1)
X_test = scaler.fit_transform(X_test)
```

```python
y_test_pred = model.predict(X_test)
```

**Converting into a CSV file**

```python
predictions =pd.DataFrame({'id':data1.id, 'label':y_test_pred})
predictions.to_csv('submission.csv', index=False)
```
